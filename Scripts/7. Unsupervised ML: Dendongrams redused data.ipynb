```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib.pyplot import figure


from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
path = '~/Desktop/CareerFoundry/3.1./'
reduced = pd.read_csv(os.path.join(path, 'Data/Clean/PCA_climate_2022.csv'))
climate = pd.read_pickle(os.path.join(path, 'Data/Clean/weather_scaled.pkl'))
```

```python
reduced.head()
```

Hierarchical Clustering Single Method Reduced Dataset (2022yr)

```python
reduced_columns = reduced.columns.drop('Unnamed: 0')
distance = linkage(
    reduced[reduced_columns],
    method = 'single'
)

plt.figure(figsize=(18,6))
dendrogram(distance, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Hierarchical Clustering Single Method Reduced Dataset (2022yr)",fontsize=18)
plt.show()
```
